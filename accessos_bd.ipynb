## Accessos a Onco_connect.db

#### Importacions i Variables Globals

```python
import sqlite3
from pathlib import Path

# Use the current working directory in Jupyter (no __file__ available)
BASE_DIR = Path.cwd()
DB_PATH = BASE_DIR / "onco_connect.db"

conn = sqlite3.connect(str(DB_PATH))
c = conn.cursor()
```

#### Taules:

```python
c.execute("SELECT name FROM sqlite_master WHERE type='table';")
tables = c.fetchall()
for table in tables:
    print(table)
```

```python
for TAULA in ['users', 'login_history', 'test_results', 'incidencies', 'daily_checkin', 'logs']: 
    c.execute(f"SELECT * FROM {TAULA};")
    rows = c.fetchall()
    print(f"\nDades de la taula {TAULA}:")
    for row in rows:
        print(row)
```

### Veure tot d'una taula :

```python
TAULA = 'logs'
c.execute(f"SELECT * FROM {TAULA};")
rows = c.fetchall()
for row in rows:
    print(row)
```
