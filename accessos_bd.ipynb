#### Importacions i Variables Globals

```python
import sqlite3
from pathlib import Path

# Use the current working directory in Jupyter (no __file__ available)
BASE_DIR = Path.cwd()
DB_PATH = BASE_DIR / "onco_connect.db"

conn = sqlite3.connect(str(DB_PATH))
c = conn.cursor()
```

#### Taules:

```python
c.execute("SELECT name FROM sqlite_master WHERE type='table';")
tables = c.fetchall()
for table in tables:
    print(table)
```
