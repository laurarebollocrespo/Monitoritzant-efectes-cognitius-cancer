```python
from datetime import datetime, timedelta
import random
# ---------------------------------------------------------
# PACIENTS 2 A 10 (Generació Automàtica amb TOTS els tests)
# ---------------------------------------------------------
tests_types = ["Fluencia", "Atencio", "Memoria", "Velocitat"]
start_date = datetime.now() - timedelta(days=60)

for i in range(2, 11):
    user = f"pacient{i}"
    print(f"Generant dades per {user}...")
    
    # Generem dades per als últims 60 dies
    for day in range(60):
        current_date = start_date + timedelta(days=day)
        date_str = current_date.strftime("%Y-%m-%d")
        datetime_str = current_date.strftime("%Y-%m-%d %H:%M:%S")
        
        # 30% de probabilitat de fer login i activitat
        if random.random() > 0.7:
            # Login
            c.execute("INSERT OR IGNORE INTO login_history (username, login_time) VALUES (?, ?)", (user, datetime_str))
            
            # Checkin
            face = random.randint(1, 5)
            c.execute("INSERT OR IGNORE INTO daily_checkin (username, date, face) VALUES (?, ?, ?)", (user, date_str, face))
            
            # Tests (TOTS ELS TIPUS DISPONIBLES)
            for test in tests_types:
                if random.random() > 0.6: # No fan tots els tests cada dia
                    score = round(random.uniform(3, 10), 2)
                    if test == "Velocitat": score = round(random.uniform(20, 60), 2) # Temps en segons
                    
                    c.execute("INSERT OR IGNORE INTO test_results (username, test_type, date, score) VALUES (?, ?, ?, ?)", 
                              (user, test, date_str, score))
            
            # Incidències (poques)
            if random.random() > 0.9:
                c.execute("INSERT OR IGNORE INTO incidencies (username, date, incidencia) VALUES (?, ?, ?)", 
                          (user, datetime_str, random.randint(0, 9)))

print("Guardant canvis...")
conn.commit()
conn.close()
print("FET! Base de dades actualitzada.")
```
